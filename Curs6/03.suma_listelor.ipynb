```python
lista_1 = [2, 3, 4, 5, 6]
lista_2 = [1, 2, 3]
```

```python
lista_rezultat = []

if len(lista_2) > len(lista_1):
    min_lista = lista_1
    max_lista = lista_2
else:
    min_lista = lista_2
    max_lista = lista_1

for i in range(len(min_lista)):
    lista_rezultat.append(lista_1[i] + lista_2[i])
    print("i=", i)



lista_rezultat = lista_rezultat + max_lista[ len(min_lista) : ]
print("lista_rezultat =", lista_rezultat)
```
