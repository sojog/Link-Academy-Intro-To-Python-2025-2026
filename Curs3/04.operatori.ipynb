# Operatori logici

AND & OR - sunt operator binari

AND -> daca ambele sunt adevarate atunci rezultatul este True, daca nu rezulatul este False

```python
print("True and False  -> R:", True and False)
```

```python
print("False and False  -> R:", False and False)
```

```python
print("False and True  -> R:", False and True)
```

```python
print("True and True  -> R:", True and True)
```

OR (sau) -> daca cel putin una este adevaratÄƒ atunci rezultatul este True, daca nu rezulatul este False

```python
print("True or False  -> R:", True or False)
print("False or False  -> R:", False or False)
print("True or True  -> R:", True or True)
print("False or  False  -> R:", False or False)
```
