```python
bancnote = []

print("Introduce bancnotele care vor fi folosite pentru calcul")

ban = input("Introdu prima bancnota")
ban = ban.strip()

while ban:  

    while not ban.isnumeric():
        ban = input("Introdu prima bancnota un numar")  
    ban = int(ban)
    bancnote.append(ban)
    print("Lungimea liste", len(bancnote))
    print("componenta listei", bancnote)
    ban = input("Introdu un numar")


```
