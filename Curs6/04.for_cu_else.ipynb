# Break - intrerupe bucla

```python
culori = ["alb", "blue", "roz", "galben", "verde"]
for i in culori:
    if i == "galben":
        break

    print("i=", i)


```

# Continue  - sare peste iteratia curenta

```python
culori = ["alb", "blue", "roz", "galben", "verde"]
for i in culori:
    if i == "galben":
        continue

    print("i=", i)
```

```python
culori = ["alb", "blue", "roz", "galben", "verde"]
for i in culori:
    if i == "galben":
        break

    print("i=", i)
else:
    print("Am ajuns la final")
```

```python
culori = ["alb", "blue", "roz", "galben", "verde"]
for i in culori:
    if i == "galben":
        break

    print("i=", i)
else:
    print("Am ajuns la final")
print("A iesit din bucla for")
```
