# Generaza un numar aleatoriu in intervalul [1, 10]

```python
random.randint(1, 10)
```

```python
random.randint(1, 200)
```

```python
calculator = random.randint(1, 10)

utilizator = input("Introdu un numar intre 1 si 10")
utilizator = utilizator.strip()

while not utilizator.isnumeric() or (int(utilizator) < 1 or int(utilizator) > 10):
    utilizator = input("Nu ai introdus un numar. Intro un numar intre 1 si 10")

utilizator = int(utilizator)
    
"Ai ghicit" if calculator == utilizator else f"Nu ai ghicit. Valoarea corecta era {calculator}"
```
