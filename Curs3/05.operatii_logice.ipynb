## Transformari in bool

```python
x = 1
x = bool(x)
print(x, type(x))
```

```python
x = 0
x = bool(x)
print(x, type(x))
```

```python
# Orice numar diferi de 0 transformat in bool este True
x = 3123
x = bool(x)
print(x, type(x))
```

```python
x = 3123
if x != 0:
    print("Numar nenul")
else:
    print("Numar nul")
```

```python
x = 0
if x != 0:
    print("Numar nenul")
else:
    print("Numar nul")
```

```python
x = 0
if x:
    print("Numar nenul")
else:
    print("Numar nul")
```

```python
x = 0
if bool(x):
    print("Numar nenul")
else:
    print("Numar nul")
```

```python
x = 312
if x:
    print("Numar nenul")
else:
    print("Numar nul")
```

# Transformare din float in bool


```python
x = 31.031
print(x, type(x))
```

# Toate numere float cu exceptia 0.0 transformate la bool sunt True. 0.0 este False

```python
x = -31.031
x = bool(x)
print(x, type(x))
```

```python
x = 0.0
x = bool(x)
print(x, type(x))
```

```python
x = 31.0
if x:
    print("Numar nenul")
else:
    print("Numar nul")
```
