```python
lista = [123, 321, 123, 5,31,32, 35, 56, 50]
odd = []
even = []
```

```python
for i in lista:
    if i % 2:  
        odd.append(i)
    else:
        even.append(i)
print("odd", odd)
print("even", even)
```

# List comprehention

```python
lista = [123, 321, 123, 5,31,32, 35, 56, 50]
odd = [ i for i in lista   if i % 2 == 1]
even = [ i  for i in lista  if i % 2 == 0]
print("odd", odd)
print("even", even)
```
